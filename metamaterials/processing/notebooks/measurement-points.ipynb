```python
import numpy as np
import csv

unitCell = []
with open('../../data/simulation/unitCell_new.csv', newline='') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    for row in data:
        unitCell.append([float(row[0]), float(row[1]), float(row[2])])
print(unitCell)

x_grid = []
y_grid = []
x_response = []
y_response = []




for i in range(14):
    for j in range(14-i):
        x_grid.append(j+0.5*i)
        y_grid.append(i)
        
        if i==0:
            x_response.append(j)
            y_response.append(i)
        
print(x_grid)
print(len(y_grid))

L = 0.008
a = np.sqrt(3)/2*L
L_real = a*(np.sqrt(3)+1)
z = 0.0015

measurePt = []
responsePt = []

for i in range(len(x_grid)):
    for point in unitCell:
        measurePt.append([point[0]+x_grid[i]*L_real*np.sqrt(3), point[1]+y_grid[i]*L_real*1.5, z])
            
for i in range(len(x_response)):
    for point in unitCell:
        responsePt.append([point[0]+x_response[i]*L_real*np.sqrt(3), point[1]+y_response[i]*L_real*1.5, z])
        
responseLine = []
for i in range(len(x_response)):
    point = unitCell[2]
    responseLine.append([point[0]+x_response[i]*L_real*np.sqrt(3), point[1]+y_response[i]*L_real*1.5, z])        
        
        
with open('../../data/simulation/fourier16.txt', 'w') as f:
    for point in measurePt:
        f.write(str(point[0]) + ',' + str(point[1]) + ',' + str(point[2]) + '\n')

with open('../../data/simulation/response.txt', 'w') as f:
    for point in responsePt:
        f.write(str(point[0]) + ',' + str(point[1]) + ',' + str(point[2]) + '\n')
        
        
with open('../../data/simulation/response_line.txt', 'w') as f:
    for point in responseLine:
        f.write(str(point[0]) + ',' + str(point[1]) + ',' + str(point[2]) + '\n')
    
print(len(measurePt))
        
        
    
    
```
