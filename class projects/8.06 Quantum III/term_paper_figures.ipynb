```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 20})

R = 1

Phi = np.linspace(-1.1, 1.1, 500) #Phi/Phi_0
V = np.abs(np.cos(np.pi*Phi)) * np.sqrt(1.01/(np.cos(np.pi*Phi))**2-1)


plt.figure(figsize=(14, 6))
plt.plot(Phi, V, linewidth=3, c='k')
plt.xlabel(r'$\Phi/\Phi_0$', fontsize=20)
plt.ylabel(r'$V/(I_c R)$', fontsize=20)
plt.savefig('V_vs_flux.pdf', bbox_inches='tight')
```

```python
M=1
Ls=1
Lj=2

Phi = np.linspace(-1.1, 1.1, 500) #Phi/Phi_0
Leff = -M**2/(Ls + Lj/np.cos(2*np.pi*Phi)) # L_eff-L_r

plt.figure(figsize=(14, 6))
plt.plot(Phi, Leff, linewidth=3, c='k')
plt.xlabel(r'$\Phi/\Phi_0$', fontsize=20)
plt.ylabel(r'$L_{eff} - L_r$ (F)', fontsize=20)
plt.savefig('L_vs_flux.pdf', bbox_inches='tight')
```
