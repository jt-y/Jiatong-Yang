```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import cycler
```

```python
energy_num=10
flux_num=10
cycle_length = 8
fluxes = np.linspace(0, 5, 5*cycle_length+1)
n = np.arange(0, energy_num,1)

color = plt.cm.viridis(np.flip(np.linspace(1, 0, cycle_length)))
mpl.rcParams['axes.prop_cycle'] = cycler.cycler('color', color)
colors = ['C'+str(i) for i in range(cycle_length)]

for flux in fluxes:
    E = (n-flux)**2
    plt.plot(flux*np.ones(energy_num), E, 'o')



plt.ylim(0, 10)
plt.xlim(0,5)
plt.xlabel(r'$\Phi/\Phi_0$')
plt.ylabel('Normalized Energy [unitless]')
plt.savefig('./hw5-plotter.pdf')
```
