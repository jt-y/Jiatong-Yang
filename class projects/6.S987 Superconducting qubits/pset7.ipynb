```python
# Author: Jiatong Yang
# Last modified: 04/06/2025

import numpy as np
import matplotlib.pyplot as plt
import scipy
from tqdm import tqdm
```

```python
wr = 2*np.pi*7e9
Z0 = 50
C = 360e-12
R = [6e6, 100e3]
Cg = 10e-12
w = np.linspace(6.85e9, 6.95e9, 5000)*2*np.pi

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6, 9))

for i in range(len(R)):
    ZL = R[i]/(1+2j*(w-wr)*C*R[i])
    Z = 1/(1j*w*Cg) + ZL
    S11 = -Z0/Z/(2+Z0/Z)
    S21 = 2/(2+Z0/Z)

    # Plot the amplitude of S11 and S21
    ax[0].plot(w/2/np.pi/1e9, 20*np.log10(np.abs(S11)), label=f'S11, R={R[i]/1e6}MOhm')
    ax[0].plot(w/2/np.pi/1e9, 20*np.log10(np.abs(S21)), label=f'S21, R={R[i]/1e6}MOhm')

    ax[1].plot(w/2/np.pi/1e9, np.angle(S11), label=f'S11,R={R[i]/1e6}MOhm')
    ax[1].plot(w/2/np.pi/1e9, np.angle(S21), label=f'S21,R={R[i]/1e6}MOhm')


ax[0].set_xlabel('Frequency (GHz)')
ax[0].set_ylabel('Magnitude (dB)')
ax[0].legend()
ax[0].grid()

ax[1].set_xlabel('Frequency (GHz)')
ax[1].set_ylabel('Phase (rad)')
ax[1].legend()
ax[1].grid()

```
