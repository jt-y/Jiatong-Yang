```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from numpy import sqrt as sqrt
```

```python
sweep_res = 100
part1 = sqrt(3)/2
part2 = 0.5
part3 = 1
sweep = np.linspace(0, part1+part2+part3, sweep_res)



k1 = np.zeros_like(sweep)
k2 = np.zeros_like(sweep)

for i in range(len(sweep)):
    if sweep[i] <= part1:
        k1[i] = sweep[i]/part1*np.pi
        k2[i] = sweep[i]/part1*np.pi

    elif sweep[i] <= part1+part2:
        k1[i] = np.pi * (1 + (sweep[i]-part1)/part2*1/3)
        k2[i] = np.pi * (1 - (sweep[i]-part1)/part2*1/3)

    else:
        k1[i] = np.pi*4/3 * (1 - (sweep[i]-(part1+part2))/part3)
        k2[i] = np.pi*2/3 * (1 - (sweep[i]-(part1+part2))/part3)

a = 3 - 2*np.cos(k1)-0.5*np.cos(k2)-0.5*np.cos(k2-k1)
b = 3 - 3/2*np.cos(k2)-3/2*np.cos(k2-k1)
c = sqrt(3)/2 * (-np.cos(k2)+np.cos(k2-k1))

omega_plus = sqrt((a+b)/2 + sqrt((a-b)**2/4+c**2))
omega_minus = sqrt((a+b)/2 - sqrt((a-b)**2/4+c**2))

plt.figure(figsize=(10,8))
plt.plot(sweep, omega_plus, '.')
plt.plot(sweep, omega_minus, '.')
plt.axvline(x=part1)
plt.axvline(x=part1+part2)
plt.xticks([0, part1, part1+part2, part1+part2+part3], ['$\Gamma$', 'M', 'R', '$\Gamma$'])
plt.ylabel('$\omega$')

plt.figure()
b1 = [2*np.pi, -2*np.pi/sqrt(3)]
b2 = [0, 4*np.pi/sqrt(3)]
kx = (k1*b1[0] + k2*b2[0])/(2*np.pi)
ky = (k1*b1[1] + k2*b2[1])/(2*np.pi)

bz_x = np.array([1, 0.5, -0.5, -1, -0.5, 0.5, 1]) * 4*np.pi/3
bz_y = np.array([0, sqrt(3)/2, sqrt(3)/2, 0, -sqrt(3)/2, -sqrt(3)/2, 0]) * 4*np.pi/3
plt.plot(kx,ky, '.')
plt.plot(bz_x, bz_y)
ax = plt.gca()
ax.set_aspect('equal')
```
