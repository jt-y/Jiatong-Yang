```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
```

### Calculating $\gamma_{11}$

- CsCl (BCC): 9.3
- ZnS (diamond): 4.07
- NaCl (SC): 6.3

```python
def create_lattice(basis, delta, unitcell_num, max_radius):
    xc = []
    yc = []
    zc = []
    rc = []
    for i in range(-unitcell_num, unitcell_num+1, 1):
        for j in range(-unitcell_num, unitcell_num+1, 1):
            for k in range(-unitcell_num, unitcell_num+1, 1):
                xc_crt = (i+delta)*basis[0][0]+(j+delta)*basis[1][0]+(k+delta)*basis[2][0]
                yc_crt = (i+delta)*basis[0][1]+(j+delta)*basis[1][1]+(k+delta)*basis[2][1]
                zc_crt = (i+delta)*basis[0][2]+(j+delta)*basis[1][2]+(k+delta)*basis[2][2]
                rc_crt = np.sqrt(xc_crt**2 + yc_crt**2 + zc_crt**2)
                if rc_crt <= max_radius and rc_crt != 0:
                    xc.append(xc_crt)
                    yc.append(yc_crt)
                    zc.append(zc_crt)
                    
    return xc, yc, zc
```

```python
# creating lattices

unitcell_num = 1 # 20 x 20 array of primitive unit cells
lattice_type = 'diamond'
max_radius = 20
n = 1

x = np.linspace(-unitcell_num, unitcell_num, 2*unitcell_num+1, endpoint=True)
y = np.linspace(-unitcell_num, unitcell_num, 2*unitcell_num+1, endpoint=True)
z = np.linspace(-unitcell_num, unitcell_num, 2*unitcell_num+1, endpoint=True)

xv, yv, zv = np.meshgrid(x, y, z)

if lattice_type == 'SC':
    a = np.array([[1,0,0], [0,1,0], [0,0,1]])
    xc, yc, zc = create_lattice(basis=a, delta = 0, unitcell_num=unitcell_num, max_radius=max_radius)

elif lattice_type == 'BCC':
    a = np.array([[1,0,0], [0,1,0], [0.5, 0.5, 0.5]])/(np.sqrt(3)/2)
    xc, yc, zc = create_lattice(basis=a, delta = 0, unitcell_num=unitcell_num, max_radius=max_radius)

elif lattice_type == 'diamond':
    a = np.array([[0.5, 0.5, 0], [0.5, 0, 0.5], [0, 0.5, 0.5]])/(np.sqrt(3)/4)
    xc1, yc1, zc1 = create_lattice(basis=a, delta = 0, unitcell_num=unitcell_num, max_radius=max_radius)
    xc2, yc2, zc2 = create_lattice(basis=a, delta = 0.25/(np.sqrt(3)/4), unitcell_num=unitcell_num, max_radius=max_radius)

    xc = xc1 + xc2
    yc = yc1 + yc2
    zc = zc1 + zc2


xc = np.array(xc)
yc = np.array(yc)
zc = np.array(zc)

```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)
ax.scatter(xc, yc, zc, c='b')
plt.show
```

```python
gamma_11 = np.sum((np.sqrt(xc**2 + yc**2 + zc**2))**(-11))

print(gamma_11)
    
```

## Finding the phase diagram

```python
lattice_type = ['NaCl (SC)', 'CsCl(BCC)', 'ZnS(Diamond)']
gamma_11 = [6.3, 9.3, 4.07]
gamma_6 = [8.04, 12.2, 5.12]
alpha = [1.747565, 1.762675, 1.6381]
```

```python
n_raw = np.linspace(0, 10, 100)
f_raw = np.linspace(0, 0.1, 100)

f, n = np.meshgrid(f_raw, n_raw)

U = []

for i in range(3):
    F = (np.sqrt((6*f/n**2*gamma_6[i]/alpha[i])**2+22*f/n**2*gamma_11[i]/alpha[i]) - 6*f/n**2*gamma_6[i]/alpha[i])**0.2
    U.append(-n**2/(2*F)+f*gamma_11[i]/F**11-f*gamma_6[i]/F**6)


phase_diag = np.argmin(U, axis=0)

plt.imshow(phase_diag, origin='lower', extent = [0, 0.1, 0, 10], aspect='auto')
plt.colorbar()
```
