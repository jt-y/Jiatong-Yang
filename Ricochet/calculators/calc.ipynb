```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S_21_min = 0 # dB
f_0 = 5e9
f = np.linspace(4.9e9, 5.1e9, 1000)

Ql_list = np.linspace(1e2, 5e3, 5)

for Ql in Ql_list:
    S_21 = (S_21_min+2j*Ql*(f-f_0)/f_0)/(1+2j*Ql*(f-f_0)/f_0)
    S_21_mag = np.abs(S_21)

    plt.plot(f/1e9, S_21_mag, label=f'$Q_l$={Ql:.0f}')
    plt.xlabel('Frequency (GHz)')
    plt.ylabel('|S21| (dB)')
    plt.legend()
```

```python
S_21_min = 0 # dB
f_0 = 5e9
f = np.linspace(4.9e9, 5.1e9, 1000)

Ql_list = np.linspace(1e2, 1e3, 5)
Qc = 1e4
phi= 0

for Ql in Ql_list:
    S_21 = 1-Ql/Qc*np.exp(complex(0,1)*phi)/(1+2j*Ql*(f-f_0)/f_0)
    S_21_mag = np.abs(S_21)

    plt.plot(f/1e9, S_21_mag, label=f'Ql={Ql:.0f}')
    plt.xlabel('Frequency (GHz)')
    plt.ylabel('|S21| (dB)')
    plt.legend()
```
